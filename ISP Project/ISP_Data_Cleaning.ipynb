```python
import pandas as pd
import numpy as np
```

```python
data_path = r"C:\Users\trush\OneDrive\Documents\WFU Grad School Info\BAN 6025 Machine Learning\Data\internet_service_churn.csv"
isp_df = pd.read_csv(data_path)
isp_df.head()
```

Data Cleaning

```python
def percentage_missing(df):
    return df.isna().sum() / len(df) * 100
```

```python
#We do not want the id column since it is not useful to determining the isp churn
isp_df = isp_df.drop(columns=['id'])
isp_df.head()
```

Missing Values

```python
percentage_missing(isp_df)
```

```python
#If a column contains more than 20% missing values, then remove the column from the model.
isp_df = isp_df.drop(columns=['reamining_contract'])
isp_df.head()
```

```python
#Remove the rest of the missing values since download_avg and upload_average have <1% missing values
isp_df = isp_df[isp_df['download_avg'].notna()]
isp_df = isp_df[isp_df['upload_avg'].notna()]
isp_df.head()
```

```python
#Everything should be good now
percentage_missing(isp_df)
```

```python
#Download as .csv
isp_df.to_csv('isp_cleaned.csv', index=False)
```
